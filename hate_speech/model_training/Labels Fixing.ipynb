```python
import pandas as pd

from pathlib import Path

from IPython.display import display
```

```python
review_path = Path('../labels/review')
raw_path = Path('../labels/raw')
```

```python
def display_full(dataframe: pd.DataFrame):
    pd.set_option("display.max_colwidth", None)
    pd.set_option("display.max_columns", None)
    display(dataframe)
    pd.reset_option("display.max_colwidth")
    pd.reset_option("display.max_columns")
```

```python
reviewed_dataframes = {}
for reviewed_file in review_path.glob("*.csv"):
    try:
        dataframe = pd.read_csv(reviewed_file, header=0, lineterminator='\n')
    except UnicodeDecodeError:
        dataframe = pd.read_csv(reviewed_file, header=0, lineterminator='\n', encoding='iso-8859-1')
    if len(dataframe[dataframe.new_labels.isna()]) > 0:
            dataframe = dataframe[~dataframe.new_labels.isna()]
    else:
        dataframe = dataframe[dataframe.new_labels != dataframe.labels]
    reviewed_dataframes[reviewed_file.name] = dataframe
```

```python
labeled_dataframes = {}
for labeled_file in raw_path.glob("*.csv"):
    try:
        dataframe = pd.read_csv(labeled_file, header=0, lineterminator='\n')
    except UnicodeDecodeError:
        dataframe = pd.read_csv(labeled_file, header=0, lineterminator='\n', encoding='iso-8859-1')
    labeled_dataframes[labeled_file.name] = dataframe
```

```python
for dataframe_name in labeled_dataframes.keys():
    print(f'Original Labels: {dataframe_name}')
    dataframe = labeled_dataframes[dataframe_name]
#     display(dataframe.head())
    for relabeled_name in sorted(reviewed_dataframes.keys(), reverse=True):
        print(f'Reviewed Labels: {relabeled_name}')
        relabeled_df = reviewed_dataframes[relabeled_name]
        merged_raw_df = relabeled_df.merge(dataframe, on=['text', 'index'])
#         print(merged_raw_df.columns)
        display_full(merged_raw_df[merged_raw_df.labels != merged_raw_df.new_labels][["index", "predicted", "labels", "new_labels", "text"]])

        for idx, row in merged_raw_df.iterrows():
            if row['new_labels'] != 3:
                if ('haaya' in relabeled_name) or ('manuel' in relabeled_name and 'haaya' not in dataframe_name):

#                 display_full(row)
#                 display_full()
                    dataframe[dataframe['index'] == row['index']].iloc[0]['class'] = row['new_labels']
    dataframe.to_csv(review_path / dataframe_name, index=None, line_terminator='\n')
```
